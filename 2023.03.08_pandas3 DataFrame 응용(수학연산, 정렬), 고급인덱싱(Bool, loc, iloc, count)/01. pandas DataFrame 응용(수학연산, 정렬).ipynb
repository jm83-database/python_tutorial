## DataFrame 수학 연산

### nunique() 메서드

```python
# nunique()는 DataFrame의 각 컬럼마다 갖는 고유 값을
# Series로 반환

titanic.nunique()
```

### count() 메서드

```python
# count()는 컬럼마다의 데이터의 개수를 계산
# NaN값은 개수에 포함시키지 않음
# 컬럼마다의 개수를 보여줘야해서 Series 객체로 값을 반환

# DataFrame 객체를 인수로 해서 len() 함수를 사용하면 
# row index의 전체 크기를 알려준다.
# NaN 값과 관계없이 전체를 세기 때문에 단 하나의 값을 정수로 반환

titanic.count()
```

```python
len(titanic)
```

### sum() 메서드

```python
# 기본적으로 NaN 값은 포함하지 않고 계산
# skipna = False로 설정하면 NaN을 포함한 결과를 얻음

# numeric_only를 적용하면 수치형 컬럼(bool 포함)만 계산해서 얻음

titanic.sum(numeric_only = True)
```

### mean(), median(), std() 메서드

```python
# 평균 : mean(), 중앙값 : median(), 표준편차 : std()

titanic.mean(numeric_only = True)
```

```python
titanic.median(numeric_only = True)
```

```python
titanic.std(numeric_only = True)
```

### max(), min() 메서드

```python
# max()는 Series에서 가장 큰값(DataFrame에서는 각 컬럼)
# min()는 Series에서 가장 작은 값

titanic.max(numeric_only = True)
```

```python
titanic.min(numeric_only = True)
```

### DataFrame 수학 연산 축의 방향

```python
# axis = 0 혹은 "rows"일 때 행간 연산
# 반환 값은 각 컬렴별 값이 나온다.
# axis의 default 값은 0

titanic[["age", "fare", "survived"]].sum(axis = "rows")
```

```python
titanic[["age", "fare", "survived"]].sum(axis = 0)
```

```python
# axis = 1 혹은 "columns"일 때 열간 연산
# 반환 값은 로우별 값이 나온다.

titanic[["age", "fare", "survived"]].sum(axis = "columns")
```

```python
titanic[["age", "fare", "survived"]].sum(axis = 1)
```

### nsmallest(), nlargest() 메서드

```python
# nsmallest()는 작은 값을 순서대로 입력한 인수의 숫자만큼 구해줌
# nlargest()는 큰 값을 순서대로 입력한 인수의 숫자만큼 구해줌

titanic.nsmallest(3, columns = ["age"])
```

```python
titanic.nlargest(3, columns = ["fare", "age"])
# 2개 열 전체에 대해서 정렬함
# 왼쪽에 있는 열부터 우선순위 적용
```

### describe() 메서드

```python
# describe()는 수치 값을 갖는 DataFrame의 각 컬럼에 대해 count, mean,
# std, min, 25%, median(50%), 75%, max에 대한 모든 통계를 구해줌

titanic.describe()
```

# DataFrame의 값 정렬하기

## sort_values() 메서드

```python
titanic.sort_values("fare") # "fare" 컬럼에 대해 값 정렬(오름차순)
```

```python
# 첫 번째 인자는 by 키워드를 가지며 정렬할 칼럼을 지정 가능
titanic.sort_values(by = ["fare", "age"]) 
# "fare", "age" 컬럼에 대해 값 정렬(오름차순)
```

```python
# 내림차순을 보고 싶다면 ascending = Fasle 설정
titanic.sort_values(by = ["fare", "age"], ascending =  False) 
```

```python
# 각 컬럼마다 정렬을 달히하고 싶으면 ascending = [True, False, False]
# 리스트 형태로 컬럼명 전달 (True는 오름차순, False는 내림차순)

# na_position = "first", "last"를 통해 결측치 정렬 방향 지정 가능

titanic.sort_values(by = ["fare", "age", "sex"], 
                   ascending = [False, True, True],
                   na_position = "last"
                  )
```

## sort_index() 메서드

```python
# sort_index는 default 값이 row index이다.
titanic.sort_index(ascending = False)
```

```python
# sort_index(axis = 1)로 지정하면 column index 기준으로 정렬한다.
titanic.sort_index(axis = 1)
```

```python
# columns index를 기준으로 정렬할 때도 ascending = False를 하면
# 내림차순 정렬 가능
titanic.sort_index(axis = 1, ascending = False)
```
