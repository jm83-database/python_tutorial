```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.dtypes
```

```python
x = ['사망자', '생존자']
y = [len(titanic.survived) - np.sum(titanic.survived), np.sum(titanic.survived)]
# number = [len(titanic.loc[titanic.survived == 0]), len(titanic.loc[titanic.survived == 1])]
colors = ["red", "green"]

plt.title("타이타닉호 생존 여부")
plt.barh(x, y, color = colors)
plt.xlabel("인원 수")
plt.show()
```

```python
bar_chart = titanic.alive.value_counts()
bar_chart.rename({'no' : '사망자', 'yes' : '생존자'}, inplace = True)
bar_chart
```

```python
alive = bar_chart.index
value = bar_chart.values
bar_colors = ["red", "green"]

plt.title("타이타닉호 생존 여부")
plt.barh(alive, value, color = bar_colors)
plt.xlabel('인원 수')
plt.show()
```

```python
import pandas as pd
titanic_age = titanic.dropna(subset = 'age').age

labels = ['미성년자', '청년', '장년', '중년', '노년']
bins = [0, 20, 30, 50, 70, 100]

sizes = pd.cut(titanic_age, bins = bins, labels = labels).value_counts(normalize = True)

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'orange']
explode = (0.1, 0, 0, 0, 0) # 카테고리간 이격도

plt.title("titanic age")
plt.pie(sizes, explode = explode, labels = sizes.index, colors = colors,
        autopct = '%1.1f%%', shadow = True, startangle = 90)
# labels = sizes.index를 통해 데이터 매칭
plt.axis('equal') # 원의 형태를 유지
plt.show()
```
