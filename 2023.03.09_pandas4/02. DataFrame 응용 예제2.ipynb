```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
# 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 column인 
# category1열을 만들어보세요.
# category1 카테고리는 다음과 같이 정의된다.

# 20살이 넘으면 성별을 그대로 사용
# 20살 미만이면 성별에 관계없이 "child"

titanic["category1"] = titanic.apply(lambda x : x.sex if x.age >= 20
                                     else "child", axis = 1)
titanic
```

```python
def adult_child(titanic):
    if titanic.age >= 20:
        return titanic.sex
    return "child"

# 함수로 필터링

titanic["category2"] = titanic.apply(adult_child, axis = 1)
titanic
```

```python
del titanic["category1"]
del titanic["category2"] # "category1", "category2"열 삭제
```

```python
# titanic에 category2 column 생성

# 조건
# (1) 성별을 나타내는 문자열 male 또는 female로 시작
# (2) 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열이 온다.
# 예를 들어 27살 남성은 male27.0값이 된다.

titanic['category2'] = titanic.sex + titanic.age.astype('str')
# sex는 str, age는 int이므로 str 형태로 변경
titanic[["age", "category2"]] # age, category2 컬럼 조회
```

```python
# 타이타닉호 승객을 '미성년자', '청년', '장년', '노년'
# 나이 그룹으로 나눈 후 각 그룹의 승객 비율 계산

bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
titanic_age = pd.cut(titanic.age, bins, labels = labels)
titanic_age
```

```python
titanic_age.value_counts()
```

```python
titanic_age_df = pd.DataFrame(titanic_age.value_counts())
# titanic_age DataFrame변환
titanic_age_df.columns = ["연령층"]
# age → 연령층 컬럼명 변경
titanic_age_df["연령비율"] = [round(i / titanic_age.count(), 2) for i in 
                      titanic_age.value_counts()]
# 연령비율 컬럼 추가(요소는 for문으로 채우기)
```

```python
# category3 정의
# (1) 20살 미만이면 성별에 관계없이 "미성년자?
# (2) 20살 이상이면 나이에 따라 '청년', '장년', '중년', '노년'을 구분하고
#     그 뒤에 성별을 나타내는 '남성', '여성'을 붙인다.

bins2 = [1,19, 30, 50, 70, 100]
labels2 = ["미성년자", "청년", "장년", "중년", "노년"]

titanic_age2 = pd.cut(titanic.age, bins2, labels = labels2)
titanic_age2
```

```python
titanic["category3"] = titanic_age2.astype(str) + ["남성" if _ == "male" else "여성" for _ in titanic.sex]
titanic # 미성년자 뒤에 성별이 붙으므로 조건 통합이 필요함(apply를 사용하자)
```

```python
titanic["연령층"] = pd.cut(titanic.age, bins = bins, labels = labels)
titanic['category4'] = titanic.apply(lambda x : "미성년자" if x.age < 20
                                     else x.연령층
                                     + ("남성" if x.sex == "male" else
                                        "여성"), axis = 1)
titanic
```
