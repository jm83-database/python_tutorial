```python
data = {
"국어": [80, 90, 70, 30],
"영어": [90, 70, 60, 40],
"수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
#(1)
print(df["수학"])
print()
#(2)
print(df[["국어","영어"]])
print()
#(3)
df["과목 평균"] = ((df["국어"] + df["영어"] + df["수학"]) / 3).round(2)
print(df)
print()
#(4)
print(df["춘향":"춘향"])
print()
#(5)
print(df.T["향단"])
# df.loc[index]를 하면 row에 대하여 Series를 뽑아내준다.
```
